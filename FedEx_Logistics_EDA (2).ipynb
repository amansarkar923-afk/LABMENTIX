#  Data Wrangling

```python
#convert Date columns
date_cols = ["PO First Sent to Client Date", "PO Sent to Vendor Date","Scheduled Delivery Date","Delivered to Client Date"]
for col in date_cols:
    if col in df.columns :
        df[col] = pd.to_datetime(df[col],errors = 'coerce')
        
```

```python
#convert Numerical Columns
num_cols = ["Line Item Quantity","Line Item Value","Pack Price","Unit Price","Weight (Kilograms)","Freight Cost (USD)" ]
for col in num_cols:
    if i in df.columns:
        df[col] = pd.to_numeric(df[col],errors = 'coerce')
```

```python
#Dosage column has 62 null values 
df["Dosage"].unique()
```

```python
#clean the column as we can see the '+'Sign
df["Dosage"].str.replace( "+", " ")
```

```python
#replacing the NaN with mode 
df['Dosage'].fillna(df['Dosage'].mode()[0], inplace=True)
```

```python
#validate that the columns has been cleaned
df.isnull().sum()
```

```python
df.info()
```

```python
#Calculate Lead Time >> Lead Time measures the time when order is placed until it is Delivered to the Customer.
df["Lead Time"] = (df["Delivered to Client Date"] - df["PO Sent to Vendor Date"]).dt.days
```

```python
#On time Delivery Flag
df["On Time"] = np.where(df["Delivered to Client Date"] < df["Scheduled Delivery Date"],1,0)

#1 → delivered before the Scheduled date
#0 → delivered on or after Scheduled date
```

```python
#Calculate Freight 
df["Freight per_kg"] =df["Freight Cost (USD)"] / df["Weight (Kilograms)"]
```

```python
#checking outliers in Numerical column
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Plot boxplots
plt.figure(figsize=(15, 10))
df[num_cols].boxplot(rot=90)

plt.title("Boxplot for Detecting Outliers")
plt.show()
```

### KPI's
